```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def LinearRegression(theta0, theta1, x, y, m,alpha,theta1_history , thet0_history):
    def h_x(x, theta0, theta1):
        return [theta0 + theta1 * xi for xi in x]

    def cost_fun(expected, y):
        errors = [expectedi - yi for expectedi, yi in zip(expected, y)]
        squared_errors = [error ** 2 for error in errors]
        sum_squared_errors = sum(squared_errors)
        cost = (1 / (2 * m)) * sum_squared_errors
        return cost 
    def gd(theta0,theta1, alpha , m ,y  ):
        expected = h_x(x, theta0, theta1)
        errors = [expectedi - yi for expectedi, yi in zip(expected, y)]
        sum_result = np.sum(errors)
        theta0 = theta0 - alpha * 1/m *sum_result
        theta1 = theta1 - alpha * 1/m *sum_result
        return [theta0 , theta1] 
    expected = h_x(x, theta0, theta1)
    cost = cost_fun(expected, y)
    for _ in range(1000):
        theta0 , theta1= gd(theta0,theta1, alpha , m  ,y )
        theta0_history.append(theta0)
        theta1_history.append(theta1)
    return cost , expected ,theta0 , theta1 , theta1_history , theta0_history

x = [1, 2, 2, 3, 4, 5, 5]
y = [7, 8, 7, 9, 11, 10, 12]
theta0 = 5  # (θ0)
theta1 = 2  # (θ1)
m = len(y)
alpha = 0.01
theta0_history = [] 
theta1_history = []
```

```python
cost , expected  ,theta0 , theta1,theta1_history , theta0_history = LinearRegression(theta0, theta1, x, y, m , alpha,theta1_history , theta0_history)
print("expected :-",expected)
print("cost" , cost)
print("theta 0 " , theta0)
print("theta 1 " , theta1 )
```
